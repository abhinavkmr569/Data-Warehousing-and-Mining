### Importing Dataset

```
df = pd.read_csv("iris.data", header = None)
df.columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Class']
df
```

```
convert_dict = {'Sepal Length': float, 'Sepal Width': float, 'Petal Length': float, 'Petal Width': float, 'Class': str}

df = df.astype(convert_dict) 
print(df.dtypes) 
```

```
train=df.sample(frac=0.67,random_state=32) 
test=df.drop(train.index)
```

### Using Test DataFrame

```
test_l = []
fin = []
test1 = test.to_numpy()

def test_list(c,k):
  for i in range(test1.shape[0]):
    for j in range(4):
      test_l.append(test1[i][j])
    #print(test_l)
    eucli_dist(test_l,c,k,fin)
    test_l.clear()
```

```
def eucli_dist(new_data_point,c,k,fin):
    eu_di = []

    for i in range(df_numpy.shape[0]):
      dist = 0.0
      for j in range(4):
        dist = dist + pow(float(df_numpy[i][j])-new_data_point[j],2)
      eu_di.append(pow(dist,0.5))

    eu_di

    train["Euclidean Distance"] = eu_di

    prediction(new_data_point,c,k,fin)
```

```
def prediction(new_data_point,c,k,fin):

  if (c!=1):
    
    train.drop(columns = ["Euclidean Distance"])

    df1 = train.sort_values(by=['Euclidean Distance']) 

    df1 = df1.iloc[:k,:]
    
    counts = dict(df1["Class"].value_counts())

    final_class = max(counts, key=counts.get) 
    print("\n\033[34m Prediction of class for data point {}: \033[1m{} \033[0m".format(new_data_point,final_class)) 

    fin.append(final_class)
    
  else:

    df1 = train.sort_values(by=['Euclidean Distance']) 

    df1 = df1.iloc[:k,:]

    counts = dict(df1["Class"].value_counts())

    final_class = max(counts, key=counts.get) 
    print("\n\033[34m Prediction of class for data point {}: \033[1m{} \033[0m".format(new_data_point,final_class)) 

    fin.append(final_class)

    train.drop(columns = ["Euclidean Distance"])
```

### Accuracy

```
def acc():
  e = list(zip(list(test["Class"]),fin))
  e1 = np.array(e)
  u = int(0)
  for i in range(e1.shape[0]):
    for j in range(e1.shape[1]):
      if (e1[i][0]!=e1[i][1]):
        u+=1
      else:
        pass

  print("\n \033[32mAccuracy: \033[1m{} % \033[0m".format(((e1.shape[0]-u)/(e1.shape[0]))*100))
```
